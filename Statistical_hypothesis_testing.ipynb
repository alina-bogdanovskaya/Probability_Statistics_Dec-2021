```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
```

Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16. Найти доверительный интервал для оценки математического ожидания с надежностью 0.95, если выборочная средняя M = 80, а объем выборки n = 256.

```python
x1 = 80 - 1.96 * 16 / math.sqrt(256)
x1
```

```python
x2 = 80 + 1.96 * 16 / math.sqrt(256)
x2
```

Доверительный интревал для оценки математического ожидания с надежностью 0.95 составляет 78.04 < m < 81.96

```python
mean = 80
sd = 16
z = 1.96

x_axis = np.arange(mean - 3.5 * sd, mean + 3.5 * sd, 0.1)

plt.rcParams["figure.figsize"] = [15, 7.50]
ax = plt.figure().add_subplot(1, 1, 1)

major_ticks_x = np.arange(mean - 3.5 * sd, mean + 3.5 * sd, 4)
major_ticks_y = np.arange(0, 0.0255, 0.0025)

ax.set_xticks(major_ticks_x)
ax.set_yticks(major_ticks_y)

ax.grid(which='both')
ax.grid(which='major', alpha=0.5)
  
y_pdf = norm.pdf(x_axis, mean, sd)

plt.plot(x_axis, y_pdf)
plt.axvline(x = mean, color = 'r', ls = '-', lw = 0.75)
plt.scatter(mean + sd * z, 0, color = 'blue')
plt.scatter(mean - sd * z, 0, color = 'blue')
plt.vlines(x = mean + sd * z, ymin = 0, ymax = 0.0125, color = 'purple', ls = '--', lw = 1)
plt.vlines(x = mean - sd * z, ymin = 0, ymax = 0.0125, color = 'purple', ls = '--', lw = 1)
plt.text(mean + sd * z, 0.0125, 'alpha/2')
plt.text(mean - sd * z, 0.0125, 'alpha/2')

plt.fill_betweenx([0, 0.0249], x1, x2, color = 'orange', alpha = 0.5)

plt.ylim(0, 0.0255)
plt.xlim(mean - 3.5*sd, mean + 3.5*sd )
```

В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные: 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1 Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

```python
measures = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]
measures.sort

mean_m = np.mean(measures)
sd = np.std(measures, ddof = 1)

print(mean_m)
print(sd)
```

```python
x1 = mean_m - 2.262 * sd / math.sqrt(10)
x1
```

```python
x2 = mean_m + 2.262 * sd / math.sqrt(10)
x2
```

С доверительной вероятностью 0.95 истинное значение X находится в интервале [6.27, 6.91]

```python
t = 2.262

x_axis = np.arange(mean_m - 3.5 * sd, mean_m + 3.5 * sd, 0.01)

plt.rcParams["figure.figsize"] = [15, 7.50]
ax = plt.figure().add_subplot(1, 1, 1)

major_ticks_x = np.arange(mean_m - 3.5 * sd, mean_m + 3.5 * sd, sd/2)
major_ticks_y = np.arange(0, 1, 0.05)

ax.set_xticks(major_ticks_x)
ax.set_yticks(major_ticks_y)

ax.grid(which='both')
ax.grid(which='major', alpha=0.5)
  
y_pdf = norm.pdf(x_axis, mean_m, sd)

plt.plot(x_axis, y_pdf)
plt.axvline(x = mean_m, color = 'r', ls = '-', lw = 0.75)
plt.scatter(mean_m + sd * t, 0, color = 'blue')
plt.scatter(mean_m - sd * t, 0, color = 'blue')
plt.vlines(x = mean_m + sd * t, ymin = 0, ymax = 0.55, color = 'purple', ls = '--', lw = 1)
plt.vlines(x = mean_m - sd * t, ymin = 0, ymax = 0.55, color = 'purple', ls = '--', lw = 1)
plt.text(mean_m + sd * t, 0.55, 'alpha/2')
plt.text(mean_m - sd * t, 0.55, 'alpha/2')

plt.fill_betweenx([0, 1], x1, x2, color = 'green', alpha = 0.5)

plt.ylim(0, 1)
plt.xlim(mean_m - 3.5*sd, mean_m + 3.5*sd )
```

Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 кв.мм.

```python
zr = (17.5 - 17)/(2/math.sqrt(100))
zr
```

На уровне значимости 0,05 гипотеза H0 не верна; средний диаметр шарика > 17 мм

```python
mean = 17
sd = 2
z = 1.645

x_axis = np.arange(mean - 3.5 * sd, mean + 3.5 * sd, 0.01)

plt.rcParams["figure.figsize"] = [15, 7.50]
ax = plt.figure().add_subplot(1, 1, 1)

major_ticks_x = np.arange(mean - 3.5 * sd, mean + 3.5 * sd, 1)
minor_ticks_x = np.arange(mean - 3.5 * sd, mean + 3.5 * sd, 0.5)
major_ticks_y = np.arange(0, 0.25, 0.05)

ax.set_xticks(major_ticks_x)
ax.set_xticks(minor_ticks_x, minor=True)
ax.set_yticks(major_ticks_y)

ax.grid(which='both')
ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)
  
y_pdf = norm.pdf(x_axis, mean, sd)

plt.plot(x_axis, y_pdf)
plt.axvline(x = mean, color = 'r', ls = '-', lw = 0.75)
plt.scatter(mean + sd * z, 0)
plt.vlines(x = mean + sd * z, ymin = 0, ymax = 0.15, color = 'purple', ls = '--', lw = 1)
plt.text(mean + sd * z, 0.15, 'alpha')
plt.scatter(mean + sd * zr, 0)
plt.vlines(x = mean + sd * zr, ymin = 0, ymax = 0.1, color = 'y', ls = '--', lw = 1.3)
plt.text(mean + sd * zr, 0.1, 'Zr')

i, = np.where(x_axis >= mean + sd * z)
fill_x = x_axis[i[0]:]
fill_y = y_pdf[i[0]:]

plt.fill_between(fill_x, fill_y, color = 'purple', alpha = 0.5)
  
plt.ylim(0, 0.25)
plt.xlim(mean - 3.5*sd, mean + 3.5*sd )
```

Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190. Известно, что их веса распределены нормально.
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%?

```python
cookies = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]
cookies.sort

mean_rep = np.mean(cookies)
sd = np.std(cookies, ddof = 1)
print(mean_rep)
print(sd)
```

```python
tr = ((mean_rep - 200) * math.sqrt(10)) / sd
tr
```

С доверительной вероятностью 99% утверждение продавца верно

```python
mean = 200
t = 3.250

x_axis = np.arange(mean - 3.5 * sd, mean + 3.5 * sd, 0.1)

plt.rcParams["figure.figsize"] = [15, 7.50]
ax = plt.figure().add_subplot(1, 1, 1)

major_ticks_x = np.arange(mean - 3.5 * sd, mean + 3.5 * sd, 1)
major_ticks_y = np.arange(0, 0.21, 0.025)

ax.set_xticks(major_ticks_x)
ax.set_yticks(major_ticks_y)

ax.grid(which='both')
ax.grid(which='major', alpha=0.5)
  
y_pdf = norm.pdf(x_axis, mean, sd)

plt.plot(x_axis, y_pdf)
plt.axvline(x = mean, color = 'r', ls = '-', lw = 0.75)
plt.scatter(mean + sd * t, 0, color = 'blue')
plt.scatter(mean - sd * t, 0, color = 'blue')
plt.vlines(x = mean + sd * t, ymin = 0, ymax = 0.15, color = 'purple', ls = '--', lw = 1)
plt.vlines(x = mean - sd * t, ymin = 0, ymax = 0.15, color = 'purple', ls = '--', lw = 1)
plt.text(mean + sd * t, 0.15, 'alpha/2')
plt.text(mean - sd * t, 0.15, 'alpha/2')
plt.scatter(mean + sd * tr, 0, color = 'orange')
plt.vlines(x = mean + sd * tr, ymin = 0, ymax = 0.075, color = 'y', ls = '--', lw = 1.3)
plt.text(mean + sd * tr, 0.075, 'Tr')

plt.ylim(0, 0.21)
plt.xlim(mean - 3.5*sd, mean + 3.5*sd )
```
